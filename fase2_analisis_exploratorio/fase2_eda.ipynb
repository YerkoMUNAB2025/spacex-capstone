```python
# Importar librer√≠as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configurar estilo de gr√°ficos
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Cargar el dataset limpio
df = pd.read_csv('../fase1_ingenieria_datos_src/spacex_launches_clean_v2.csv')

# Vista r√°pida
df.head()
```

```python
# Informaci√≥n general
df.info()

# Estad√≠sticas descriptivas
df.describe()

# Verificar valores nulos
df.isnull().sum()
```

```python
# Distribuci√≥n de la masa del payload
sns.histplot(df['payload_mass_kg'], bins=20, kde=True)
plt.title('Distribuci√≥n de la masa del payload')
plt.xlabel('Masa (kg)')
plt.ylabel('Frecuencia')
plt.show()

# Conteo de sitios de lanzamiento
sns.countplot(y='launchpad', data=df)
plt.title('Frecuencia por sitio de lanzamiento')
plt.xlabel('Cantidad de lanzamientos')
plt.ylabel('Launchpad ID')
plt.show()

# Tipos de cohetes
sns.countplot(y='rocket', data=df)
plt.title('Tipos de cohetes utilizados')
plt.xlabel('Cantidad')
plt.ylabel('Rocket ID')
plt.show()
```

```python
# Convertir landing_success a string para que Seaborn lo trate como categor√≠a
df['landing_success'] = df['landing_success'].astype(str)

# Boxplot: relaci√≥n entre masa del payload y √©xito del aterrizaje
sns.boxplot(x='landing_success', y='payload_mass_kg', data=df)
plt.title('Masa del payload seg√∫n resultado de aterrizaje')
plt.xlabel('Landing Success (1 = √©xito, 0 = fallo)')
plt.ylabel('Payload Mass (kg)')
plt.show()
```
