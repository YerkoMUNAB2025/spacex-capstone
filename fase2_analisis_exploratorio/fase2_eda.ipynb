```python
# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configurar estilo de gráficos
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Cargar el dataset limpio
df = pd.read_csv('../fase1_ingenieria_datos_src/spacex_launches_clean_v2.csv')

# Vista rápida
df.head()
```

```python
# Información general
df.info()

# Estadísticas descriptivas
df.describe()

# Verificar valores nulos
df.isnull().sum()
```

```python
# Distribución de la masa del payload
sns.histplot(df['payload_mass_kg'], bins=20, kde=True)
plt.title('Distribución de la masa del payload')
plt.xlabel('Masa (kg)')
plt.ylabel('Frecuencia')
plt.show()

# Conteo de sitios de lanzamiento
sns.countplot(y='launchpad', data=df)
plt.title('Frecuencia por sitio de lanzamiento')
plt.xlabel('Cantidad de lanzamientos')
plt.ylabel('Launchpad ID')
plt.show()

# Tipos de cohetes
sns.countplot(y='rocket', data=df)
plt.title('Tipos de cohetes utilizados')
plt.xlabel('Cantidad')
plt.ylabel('Rocket ID')
plt.show()
```

```python
# Convertir landing_success a string para que Seaborn lo trate como categoría
df['landing_success'] = df['landing_success'].astype(str)

# Boxplot: relación entre masa del payload y éxito del aterrizaje
sns.boxplot(x='landing_success', y='payload_mass_kg', data=df)
plt.title('Masa del payload según resultado de aterrizaje')
plt.xlabel('Landing Success (1 = éxito, 0 = fallo)')
plt.ylabel('Payload Mass (kg)')
plt.show()
```
